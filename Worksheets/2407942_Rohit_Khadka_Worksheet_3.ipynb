```python
#3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:
#1 Total fare in the titanic dataset
totalFare = round(titanic_df['Fare'].sum(), 2)
print(f"The total fare paid is: {totalFare}")
```

```python
#2
#Total fare of passengers in the first class

fareSumClass1 = round(titanic_df.loc[titanic_df['Pclass'] == 1, 'Fare'].sum(), 2)
print(fareSumClass1)
```

```python
#3 
#Total fare of passengers in the third and second class
fareSumClass2 = round(titanic_df.loc[(titanic_df['Pclass'] == 2), 'Fare'].sum(), 2)
fareSumClass3 = round(titanic_df.loc[(titanic_df['Pclass'] == 3), 'Fare'].sum(), 2)
print(fareSumClass2, fareSumClass3)
```

```python
#4 
#Combining the results into a list
fare_list = []
fare_list.append(float(fareSumClass1))
fare_list.append(float(fareSumClass2))
fare_list.append(float(fareSumClass3))
fare_list
```

```python
#5
#Proportion of fare revenue by class
fare_per_class = fare_list / totalFare
fare_per_class
```

```python
#What percent of the total number of passengers on the Titanic belonged to each age group (e.g., child, adult, senior)?
#1
#First drop null values in age group 
titanic_df.dropna(subset= ['Age'], inplace = True)
#Grouping into age group
titanic_df['age_group'] = pd.cut(titanic_df['Age'], bins = [0,17, 64, float('inf')], labels = ['child', 'adult', 'senior'])
# titanic_df['age_group'].sample(n = 10)
titanic_df['age_group'].value_counts()
```

```python
#2 Calculating total number of passengers in titanic
totalPassengers = titanic_df['PassengerId'].nunique()
print(f"The total number of passengers in titanic is: {totalPassengers}")
```

```python
#3 Number of passengers in each age group 
age_group_count = titanic_df['age_group'].value_counts()
age_group_count
```

```python
#4 Proportion of each age group to the total number of passengers 
age_group_ratio = age_group_count / totalPassengers
age_group_ratio
```

```python
#Displaying proportion as the percentage 
age_group_percentage = age_group_ratio * 100
age_group_percentage
```
