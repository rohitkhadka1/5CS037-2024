```python
#Exercise 8.1.1 Recursion
#Task 1
def sum_nested_list(nested_list):
  """
  Calculate the sum of all numbers in a nested list.
  This function takes a list that may contain integers and other nested lists.
  It recursively traverses the list and sums all the integers, no matter how deeply
  nested they are.
  Args:
  nested_list (list): A list that may contain integers or other lists of integers.
  Returns:
  18
  5CS037 Worksheet - 0 Siman Giri
  int: The total sum of all integers in the nested list, including those in sublists
  .
  Example:
  >>> sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]])
  36
  >>> sum_nested_list([1, [2, 3], [4, [5]]])
  15
  """
  total = 0
  for element in nested_list:
    if isinstance(element, list):
      total += sum_nested_list(element)
    else:
      total += element
  return total
nested_list =  [1, [2, [3, 4], 5], 6, [7, 8]]
print(sum_nested_list(nested_list))
```

```python
#Task 2
def generate_permutations(s):
  #Base case: if string is empty or single character
  if len(s) <= 1:
    return [s]

  #To store all possible permutations
  permutations = []

  #To track used characters
  used_chars = set()

  #To generate permutations recursively
  for i in range(len(s)):
    if s[i] in used_chars:
      continue

    used_chars.add(s[i])


  #Current character
    current_char = s[i]

  #Remaining characters
    remaining_chars = s[:i] + s[i+1:]

    for sub_perm in generate_permutations(remaining_chars):
      permutations.append(current_char + sub_perm)

  return permutations


#Test Cases
print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
#8.2.2
#Task 2
def min_coins(coins, amount):
  """
  Finds the minimum number of coins needed to make up a given amount using dynamic
  programming.
  This function solves the coin change problem by determining the fewest number of
  coins from a given set of coin denominations that sum up to a target amount. The
  solution uses dynamic programming(tabulation) to iteratively build up the minimum
  number of coins required for each amount.
  Parameters:
  coins (list of int): A list of coin denominations available for making change. Each
  coin denomination is a positive integer.
  amount (int): The target amount for which we need to find the minimum number of coins
  . It must be a non-negative integer.
  Returns:
  int: The minimum number of coins required to make the given amount.
  If it is not possible to make the amount with the given coins, returns -1.
  Example:
  >>> min_coins([1, 2, 5], 11)
  3
  >>> min_coins([2], 3)
  -1
  """
  dp = [float('inf')] * (amount + 1)
  dp[0] = 0
  for coin in coins:
   for i in range(coin, amount + 1):
     dp[i] = min(dp[i], dp[i - coin] + 1)
  return dp[amount] if dp[amount] != float('inf') else -1
print(min_coins([1, 2, 5], 11))
print(min_coins([2], 3))
```

```python
#Task 2
def longest_common_subsequence(s1, s2):
    # Get lengths of both strings
    m, n = len(s1), len(s2)

    # Create a DP table of size (m+1) x (n+1)
    # Add 1 to handle empty string case
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Build the dp table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                # If characters match, add 1 to diagonal value
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                # If characters don't match, take max of left or up
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    # Return length of LCS
    return dp[m][n]

# Let's create an enhanced version that also returns the LCS
def longest_common_subsequence_with_sequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill the dp table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    # Backtrack to find the actual subsequence
    lcs = []
    i, j = m, n
    while i > 0 and j > 0:
        if s1[i-1] == s2[j-1]:
            lcs.append(s1[i-1])
            i -= 1
            j -= 1
        elif dp[i-1][j] > dp[i][j-1]:
            i -= 1
        else:
            j -= 1

    # Return both length and sequence (reversed since we backtracked)
    return dp[m][n], ''.join(reversed(lcs))

# Test cases
test_cases = [
    ("abcde", "ace"),      # LCS = "ace", length = 3
    ("hello", "world"),    # LCS = "lo", length = 2
    ("AGGTAB", "GXTXAYB"), # LCS = "GTAB", length = 4
    ("", "abc"),          # LCS = "", length = 0
    ("abc", "abc")        # LCS = "abc", length = 3
]

# Test the first function (length only)
print("Testing LCS length:")
for s1, s2 in test_cases:
    length = longest_common_subsequence(s1, s2)
    print(f"Strings: '{s1}' and '{s2}'")
    print(f"LCS Length: {length}\n")

# Test the enhanced function (length and sequence)
print("\nTesting LCS length and sequence:")
for s1, s2 in test_cases:
    length, sequence = longest_common_subsequence_with_sequence(s1, s2)
    print(f"Strings: '{s1}' and '{s2}'")
    print(f"LCS Length: {length}")
    print(f"LCS Sequence: '{sequence}'\n")
```

```python
#Task 3
def knapsack(weights, values, capacity):
    n = len(weights)
    # Create a 2D DP table with dimensions (n+1) x (capacity+1)
    # dp[i][w] represents the maximum value that can be achieved
    # using first i items and weight capacity w
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    # Build table dp[][] in bottom up manner
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i-1] <= w:
                # If we can include the current item, take maximum of:
                # 1. Value of current item + value of remaining capacity using previous items
                # 2. Value without using current item
                dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]],
                             dp[i-1][w])
            else:
                # If current item is too heavy, don't include it
                dp[i][w] = dp[i-1][w]

    # Return the maximum value that can be achieved
    return dp[n][capacity]

# Test the function
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

result = knapsack(weights, values, capacity)
print(f"Maximum value: {result}")  # Should print 9

# Let's also add a function to track which items were selected
def knapsack_with_items(weights, values, capacity):
    n = len(weights)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    # Fill the dp table
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(values[i-1] + dp[i-1][w-weights[i-1]],
                             dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]

    # Backtrack to find selected items
    selected_items = []
    w = capacity
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i-1][w]:
            selected_items.append(i-1)
            w -= weights[i-1]

    return dp[n][capacity], selected_items

# Test with item tracking
max_value, selected = knapsack_with_items(weights, values, capacity)
print(f"Maximum value: {max_value}")
print(f"Selected items (indices): {selected}")
print("Selected items (weight, value):")
for idx in selected:
    print(f"Item {idx}: weight = {weights[idx]}, value = {values[idx]}")

#Test Cases
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
```
