```python
#Problem 2 - Data Imputations:

df2 = pd.read_csv("medical_students_dataset.csv")
df2.head(10)
```

```python
df2.tail(10)
```

```python
df2.isnull().sum()
```

```python
df2['Student ID'].fillna(method = 'ffill')
```

```python
df2['Gender'] = df2['Gender'].fillna(df2['Gender'].mode()[0])
```

```python
df2['Blood Type'] = df2['Blood Type'].fillna(df2['Blood Type'].mode()[0])
```

```python
df2['Age'] = df2['Age'].fillna(df2['Age'].median())
df2['Height'] = df2['Height'].fillna(df2['Height'].mean())
```

```python
df2['Temperature'] = df2['Temperature'].fillna(df2['Temperature'].mean())
```

```python
df2['Heart Rate'] = df2['Heart Rate'].fillna(df2['Heart Rate'].median())
```
