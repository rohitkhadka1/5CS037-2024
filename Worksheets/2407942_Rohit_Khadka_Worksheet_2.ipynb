```python
#Problem 2 - Data Imputations:

df2 = pd.read_csv("medical_students_dataset.csv")
df2.head(10)
```

```python
df2.tail(10)
```

```python
df2.isnull().sum()
```

```python
df2.info()
```

```python
df2['Student ID'] = df2['Student ID'].fillna('UNKNOWN')
```

```python
# Fill missing values in categorical columns with the mode
for col in df2.drop('Student ID', axis = 1).select_dtypes(include='object').columns:
    mode_value = df2[col].mode()[0]
    df2[col] = df2[col].fillna(mode_value)
```

```python
# Fill missing values in numeric columns with the mean
for col in df2.select_dtypes(include='number').columns:
    mean_value = df2[col].mean()
    df2[col] = df2[col].fillna(mean_value)
```

```python
df2.isnull().sum()
```

```python
print(f"The total number of duplicate entries: {df2.duplicated().sum()}")
#Viewing the rows containing duplicate rows
duplicated_rows = df2[df2.duplicated()]
duplicated_rows
      
```

```python
#Cleaning the duplicated entries
df2_duplicates_cleaned = df2.drop_duplicates(keep = 'first')
# Reset the index after dropping duplicates (optional)
df2_duplicates_cleaned.reset_index(drop=True, inplace=True)
df2_duplicates_cleaned.shape
```

```python
#Problem 3

titanic_df = pd.read_csv("Titanic-Dataset.csv")
titanic_df.head()
```

```python
titanic_df.info()
```

```python
titanic_df.describe()
```

```python
#Subsetting the dataframe
df = titanic_df[titanic_df["Pclass"] == 1][["Name", "Pclass", "Sex", "Age", "Fare", "Survived"]]
df.head()
```

```python
nullAge = df['Age'].isnull().sum()
print(f"The number of null values in Age column is: {nullAge}")
# Drop rows with null values in the 'Age' column
df_cleaned = df.dropna(subset=['Age'])

print("Rows with null 'Age' values have been dropped.")
print(f"New number of rows in DataFrame: {df_cleaned.shape[0]}")
```

```python
describe_fare = df['Fare'].describe()
describe_fare
```

```python
print(f"Mean is {describe_fare['mean']}.\nmaximum is {describe_fare['max']}.\nminimum is {describe_fare['min']}.\nand median is {describe_fare['50%']} ")
```

```python
dummies = pd.get_dummies(titanic_df["Embarked"])
titanic_df = pd.concat((dummies, titanic_df), axis = 1)
```

```python
#Checking the dataframe after encoding
titanic_df.head()
```

```python
#Dropping the encoded feature
titanic_df.drop("Embarked", axis = 1)
```

```python
#Checking the survival rate on the basis of Sex column
grouping_Sex = df.groupby('Sex')['Survived'].mean()
grouping_Sex
```
